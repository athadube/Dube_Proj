# Uber Data Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.figure as figure
%matplotlib inline
```

```python
df = pd.read_csv('uber_apr14.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date/Time'])
df.head()
```

```python
#df.dtypes
```

```python
df['hour'] = df['Date'].dt.hour
df.head()
```

```python
df['day'] = df['Date'].dt.day
df.head()
```

```python
df['month'] = df['Date'].dt.month
df.head()
```

```python
df['weekday'] = df['Date'].dt.weekday
df.head()
```

### Analyse the day of month

```python
plt.hist(df.day, bins=31, rwidth=8, range=(0.5, 30.5))
plt.xlabel('Day of the month')
plt.ylabel('Frequency')
plt.title('Frequency by Day of the Month -Uber - Apr - 2014')
plt.show()
```

```python
# Alternatively this can be achieved by grouping by day.
result1 = df.groupby('day').count()
result1
```

```python
plt.bar(result1.index, result1['Base'])
plt.xticks(result1.index, rotation='vertical', size=8)
plt.title('Frequency by day of the month')
plt.ylabel('Frequency')
plt.xlabel('days')
plt.show()
```

### Analyse by Hour

```python
result2 = df.groupby('hour').count()
result2
```

```python
plt.bar(result2.index, result2.Base)
plt.xticks(result2.index)
plt.title('Frequency by hour of the day')
plt.ylabel('Frequency')
plt.xlabel('Hours')
plt.show()
```


### Analyse by weekday

```python
result3 = df.groupby('weekday').count()
result3
```

```python
plt.bar(result3.index, result3.Base)
plt.xticks(result3.index)
plt.title('Frequency by day of the week')
plt.ylabel('Frequency')
plt.xlabel('weekdays')
plt.show()
```

### Cross analysis of Hour & Day

```python
result4 = df.groupby(['weekday', 'hour']).count()
result4
```

### Analyse by Lat and Lon

```python
plt.hist(df.Lat, bins=100, range=(40.5, 41))
plt.show()
```

```python
plt.hist(df.Lon, bins=100,  range=(-74.1, -73.9))
plt.show()
```

```python
plt.hist(df.Lat, bins=100, range=(40.5, 41), color='r', alpha=0.5, label='Latitude')
plt.legend(loc='upper left')
plt.twiny()
plt.hist(df.Lon, bins=100,  range=(-74.1, -73.9), color='g', alpha=0.5, label='Longitude')
plt.legend(loc='upper right')
plt.show()
```

```python
#plotting the Lat against Long could show the map of the area
plt.figure(figsize=(15,20))
plt.plot(df.Lon, df.Lat, ".", alpha=0.5, ms=1)
plt.xlim(-74.1, -73.9)
plt.ylim(40.5, 41)
plt.show()
```
